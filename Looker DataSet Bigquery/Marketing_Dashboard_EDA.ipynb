```python
inventory_rate = inventory_items.groupby(['product_id']).count()[['created_at', 'sold_at']]
inventory_rate = inventory_rate.rename(columns={'created_at':'warehousinng', 'sold_at':'delivery'})
inventory_rate = inventory_rate.assign(stock=inventory_rate['warehousinng'] - inventory_rate['delivery'])
inventory_rate = inventory_rate.assign(stock_rate=inventory_rate['stock'] / inventory_rate['warehousinng']*100)
inventory_rate
```

```python
inventory_rate2 = inventory_items.groupby(['product_category']).count()[['created_at', 'sold_at']]
inventory_rate2 = inventory_rate2.rename(columns={'created_at':'warehousinng', 'sold_at':'delivery'})
inventory_rate2 = inventory_rate2.assign(stock=inventory_rate2['warehousinng'] - inventory_rate2['delivery'])
inventory_rate2 = inventory_rate2.assign(stock_rate=inventory_rate2['stock'] / inventory_rate2['warehousinng']*100)
inventory_rate2
```

```python
inventory_items_cp = inventory_items.copy()
create_date_list = inventory_items_cp['created_at'].str.split('-')

inventory_items_cp = inventory_items_cp.assign(c_year=create_date_list.str.get(0).astype(int))
inventory_items_cp = inventory_items_cp.assign(c_month=create_date_list.str.get(1).astype(int))
inventory_items_cp
```

```python
inventory_turn_rate = inventory_items.groupby(['product_category']).count()[['created_at', 'sold_at']]
inventory_turn_rate = inventory_turn_rate.rename(columns={'created_at':'warehousinng', 'sold_at':'delivery'})
inventory_turn_rate = inventory_turn_rate.assign(stock=inventory_turn_rate['warehousinng'] - inventory_turn_rate['delivery'])
inventory_turn_rate = inventory_turn_rate.assign(stock_rate=inventory_turn_rate['stock'] / inventory_turn_rate['warehousinng']*100)
inventory_turn_rate = inventory_turn_rate.assign(s_year=2018)
inventory_turn_rate
```

```python
inventory_turn_rate = inventory_items.groupby(['product_category']).count()[['created_at', 'sold_at']]
inventory_turn_rate = inventory_turn_rate.rename(columns={'created_at':'warehousinng', 'sold_at':'delivery'})
inventory_turn_rate = inventory_turn_rate.assign(stock=inventory_turn_rate['warehousinng'] - inventory_turn_rate['delivery'])
inventory_turn_rate
```

```python
temp_df = pd.DataFrame(columns=inventory_items_cp.columns)
ITR_df_list = []
ITR_list = []

for key, group in inventory_items_cp.groupby('c_year') : 
    if key == inventory_items_cp['c_year'].min() : 
        temp_df = temp_df.append(group)
        continue
    inventory_last = temp_df.groupby(['product_category']).count()[['created_at', 'sold_at']]
    inventory_last = inventory_last.rename(columns={'created_at':'warehousinng', 'sold_at':'delivery'})
    inventory_last = inventory_last.assign(stock=inventory_last['warehousinng'] - inventory_last['delivery'])
    
    inventory_present = group.groupby(['product_category']).count()[['created_at', 'sold_at']]
    inventory_present = inventory_present.rename(columns={'created_at':'warehousinng', 'sold_at':'delivery'})
    inventory_present = inventory_present.assign(stock=inventory_present['warehousinng'] - inventory_present['delivery'])
    temp_df = temp_df.append(group)
    
    itr = pd.DataFrame(inventory_present['delivery'] / ((inventory_last['stock'] + inventory_present['stock']) / 2))
    ITR_list.append(itr[0].mean())
    itr = itr.rename(columns={0:f'itr_{key}'})

    ITR_df_list.append(itr)
```

```python
ITR = pd.concat(ITR_df_list, axis=1)
ITR
```
