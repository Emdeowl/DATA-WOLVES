```python
import pandas as pd
import numpy as np
import random

import seaborn as sns
import matplotlib.pyplot as plt

from tqdm import tqdm
import random
from collections import Counter
```

```python
zomato = pd.read_csv('./data/zomato_preprocessed.csv')
zomato = zomato.assign(order_time=pd.to_datetime(zomato['order_time']))
zomato
```

```python
# 광고 수익 -> 어떻게 난수를 뿌릴 것인가
# 꽤나 큰 비중으로? 76%가 광고 수익에서 나오는데

# RFM, 6개월의 이력으로 계산 
# -> R <- 최근 Success 구매일로부터 현재까지의 날짜 차이
 
# 쿠폰 수수료 ? 제할 것인가 ?
```

```python
zomato_gold = zomato[zomato['is_gold']==True].reset_index(drop=True)
zomato_regular = zomato[zomato['is_gold']==False].reset_index(drop=True)
```

```python
# commercial value
# 광고수익은 노출 단위 횟수 당 광고주가 지불하는 비용
# 어느 정도 크기의 난수를 줄 것인가는 해보고 판단
```

```python
#광고가 10개 정도 있다고 가정

weight = [0.1, 0.02, 0.15, 0.25, 0.15, 0.1, 0.1, 0.06, 0.04, 0.03]

commercial = [[5, 6], [9, 10], [6, 8], 
              [5, 6], [5, 7], [6, 8], [5, 7],
              [7, 8], [8, 9], [7, 10]]
```

```python
commercial_fee_range = commercial[random.randint(0, 9)]
print(round(random.uniform(commercial_fee_range[0], commercial_fee_range[1]), 2))
```

```python
zomato_gold = zomato_gold.assign(ad_revenue=np.NaN)
zomato_gold
```

```python
for i in tqdm(range(zomato_regular.shape[0])) : 
    commercial_fee_range = commercial[random.randint(0, 9)]
    commercial_fee = round(random.uniform(commercial_fee_range[0], commercial_fee_range[1]), 2)
    zomato_regular.loc[zomato_regular.index==i, 'ad_revenue'] = commercial_fee
```

```python
zomato_regular
```

```python
zomato_concat = pd.concat([zomato_gold, zomato_regular], axis=0)
zomato_concat = zomato_concat.sort_values('order_time').reset_index(drop=True)
zomato_concat
```

```python
zomato_concat.to_csv('./data/zomato_preprocessed_2.csv', index=False)
```
