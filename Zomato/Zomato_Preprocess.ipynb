# Are you Gold?

```python
zomato.analytical_customer_id.nunique()
#13559명의 고객 데이터
```

```python
def is_gold_true(df, iteration) : 
    global is_gold
    if df.iloc[iteration].is_affordable_gem==True : 
        is_gold = True
    else :
        pass
    
def is_gold_reset() : 
    global is_gold
    is_gold=False
```

```python
customer_list = []
is_gold = False

for _, group in tqdm(zomato.groupby('analytical_customer_id')) : 
    is_gold_reset()
    temp = group.sort_values('order_time').reset_index(drop=True)
    for i in range(temp.shape[0]) : 
        is_gold_true(temp, i)
        temp.loc[temp.index==i, 'is_gold'] = is_gold
    customer_list.append(temp)
```

```python
zomato_is_gold = pd.concat(customer_list, axis=0).sort_values('order_time').reset_index(drop=True)
zomato_is_gold
```

```python
zomato_gold = zomato_is_gold[zomato_is_gold['is_gold']==True]
zomato_not_gold = zomato_is_gold[zomato_is_gold['is_gold']==False]
```

# Vertical

```python
menu = ['Pizza', 'Desserts', 'Chinese', 'Indian','Burger', 'American', 'Italian', 'Mexican', 'Thai', 'Vietnamese']
weight = [0.1, 0.02, 0.15, 0.25, 0.15, 0.1, 0.1, 0.06, 0.04, 0.03]
```

```python
zomato_food = zomato_is_gold[zomato_is_gold['vertical_class']=='food'].reset_index(drop=True)
zomato_non_food = zomato_is_gold[zomato_is_gold['vertical_class']=='non-food'].reset_index(drop=True)
```

```python
for i in tqdm(range(zomato_food.shape[0])) : 
    zomato_food.loc[zomato_food.index==i, 'vertical'] = menu[random.choices(range(0, 10), weights=weight)[0]]
```

```python
zomato_vertical = pd.concat([zomato_food, zomato_non_food]).reset_index(drop=True).sort_values('order_time').reset_index(drop=True)
```

```python
zomato_vertical = zomato_vertical.assign(order_delay=zomato_vertical['actual_delivery_time']-zomato_vertical['promised_delivery_time'])
zomato_vertical = zomato_vertical.assign(dropoff_distance_manhattan=zomato_vertical['dropoff_distance_manhattan']/1000)
zomato_vertical = zomato_vertical.assign(promised_delivery_time=zomato_vertical['promised_delivery_time']/60)
zomato_vertical = zomato_vertical.assign(order_delay=zomato_vertical['order_delay']/60)
zomato_vertical = zomato_vertical.assign(actual_delivery_time=zomato_vertical['actual_delivery_time']/60)
```

```python
longdi_index = zomato_vertical[zomato_vertical['dropoff_distance_manhattan']>100].index
zomato_vertical = zomato_vertical.drop(longdi_index).reset_index(drop=True)
zomato_vertical
```

```python
zomato_success = zomato_vertical[(zomato_vertical['is_successful']==True)]
zomato_fail = zomato_vertical[(zomato_vertical['is_successful']==False)]
```

```python
customer_success = zomato_success.groupby('analytical_customer_id').agg({'is_gold':'max', 'basket_amount_lc':'sum', 'affordability_amt_total':'sum', 'order_id':'count'}).reset_index()
customer_success
```

```python
customer_fail = zomato_fail.groupby('analytical_customer_id').agg({'is_gold':'max', 'basket_amount_lc':'sum', 'affordability_amt_total':'sum', 'order_id':'count'}).reset_index()
customer_fail['basket_amount_lc'] = 0
customer_fail['affordability_amt_total'] = 0
customer_fail['order_id'] = 0
customer_fail
```

```python
customer = pd.concat([customer_success, customer_fail])
customer = customer.drop_duplicates('analytical_customer_id', keep='first').reset_index(drop=True)
customer = customer.rename(columns={'basket_amount_lc':'amt_total', 'affordability_amt_total':'afford_amt_total', 'order_id':'order_count'})
customer
```

```python
customer.to_csv('./data/customer.csv', index=False)
zomato_vertical.to_csv('./data/zomato_preprocessed.csv', index=False)
```
